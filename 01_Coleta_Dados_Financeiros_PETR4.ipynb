```python
# ==============================================================================
# PASSO 2: INSTALAÇÃO E IMPORTAÇÃO DE BIBLIOTECAS
# ==============================================================================
!pip install yfinance --upgrade --quiet
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
# ==============================================================================
# PASSO 3: EXTRAÇÃO E TRANSFORMAÇÃO DOS DADOS
# ==============================================================================
ticker = "PETR4.SA"
data_inicio = "2018-01-01"
data_fim = "2023-12-31"

print(f"Extraindo dados diários de {ticker} da B3...")
ativo = yf.Ticker(ticker)
df_petr4 = ativo.history(start=data_inicio, end=data_fim)

# Cálculo do Log-Retorno para garantir estacionariedade (Exigência do GARCH/XGBoost)
df_petr4['Log_Retorno'] = np.log(df_petr4['Close'] / df_petr4['Close'].shift(1))
df_petr4.dropna(inplace=True)

# Mantendo apenas as colunas essenciais
colunas_para_manter = ['Open', 'High', 'Low', 'Close', 'Volume', 'Log_Retorno']
df_petr4 = df_petr4[colunas_para_manter]
```

```python
# ==============================================================================
# PASSO 4: SALVAMENTO AUTOMATIZADO NO GOOGLE DRIVE
# ==============================================================================
caminho_arquivo_financeiro = caminho_base + "base_financeira_petr4.csv"
df_petr4.to_csv(caminho_arquivo_financeiro)

print(f"\nSucesso! Arquivo financeiro salvo permanentemente em: {caminho_arquivo_financeiro}")
display(df_petr4.head())
```
